```python
geometry_list = []
country_id = []
for x in pacific_islands['geometry']:
#     print(list(x.exterior.coords))
#     print(x.equals(geometry.Polygon(list(x.exterior.coords))))

    
    
    if str(type(x)) == "<class 'shapely.geometry.multipolygon.MultiPolygon'>":
        point_list = []
        multi_list = []
        for i in x.geoms:
            for point in list(i.exterior.coords):
                point_list.append(geometry.Point(point[0]-360,point[1]))
            multi_list.append(geometry.Polygon(point_list))
        geometry_list.append(geometry.MultiPolygon(multi_list))
            
    else:    
        point_list = []
        for point in list(x.exterior.coords):
            point_list.append(geometry.Point(point[0]-360,point[1]))  
    
        geometry_list.append(geometry.Polygon(point_list))
```
