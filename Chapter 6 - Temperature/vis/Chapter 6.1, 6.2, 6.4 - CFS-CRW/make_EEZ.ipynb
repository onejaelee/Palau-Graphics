```python
geometry_list = []

for x in all_pac_islands_neg['geometry']:

    if str(type(x)) == "<class 'shapely.geometry.multipolygon.MultiPolygon'>":
        point_list = []
        multi_list = []
        for i in x.geoms:
            for point in list(i.exterior.coords):
                if point[0] < 0:
                    x_value = point[0] + 360
                    while(x_value < 0):
                        x_value += 360
                    point_list.append(geometry.Point(x_value,point[1]))
                else:
                    point_list.append(geometry.Point(point[0],point[1]))   
            multi_list.append(geometry.Polygon(point_list))
        geometry_list.append(geometry.MultiPolygon(multi_list))
            
    else:
        point_list = []
        for point in list(x.exterior.coords):
            if point[0] < 0:
                x_value = point[0] + 360
                while(x_value < 0):
                    x_value += 360
                point_list.append(geometry.Point(x_value,point[1])) 
            else:
                point_list.append(geometry.Point(point[0],point[1]))   

        geometry_list.append(geometry.Polygon(point_list))
```
